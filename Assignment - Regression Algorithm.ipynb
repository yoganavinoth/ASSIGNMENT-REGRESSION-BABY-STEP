```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'charges', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
#sklearn library used to split train and test model
from sklearn.model_selection import train_test_split
#Then call train_test_split function it is parameterised function
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="rbf",C=3000)
regressor.fit(X_train,Y_train)
```

```python
regressor.intercept_
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
#to find r2_score,we take actual output and predictor.
r_score=r2_score(Y_test,y_pred)
```

```python
r_score
```

```python
#pickle is used to save model creation
import pickle
#Create filename,it is pickle extension so we save .sav
filename="finalized_model_supportvectormachine.sav"
```

```python
pickle.dump(regressor, open(filename,"wb"))
```

```python
#load the model and rb is used for just read 
loaded_model=pickle.load(open("finalized_model_supportvectormachine.sav","rb"))
#we check it and for prediction we can do it
result=loaded_model.predict([[18,39,52,45,65,70]])
```
